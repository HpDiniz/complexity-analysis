<h3>Hamiltonian Path Problem</h3>
<p> A Hamiltonian path, also called a Hamilton path, is a graph path between two vertices of a graph that visits each vertex exactly once. <br>
So the problem is: Given a graph G, is there a Hamiltonian path ?
<p>
<strong>The Problem is in NP?</strong><br>
    <p>To be in NP means that, given a proposed solution (a certificate) to the problem, we can verify the solution in polynomial time. In this case, that means that given a directed graph G and path between 2 vertices, we can check in polynomial time if the path is a Hamiltonian Path.</p>
   <p>If we are given a certificate, we can do the following to algorithmically check for a Hamiltonian Path:</p>
   <p><em>On input G = (V,E) where G is a directed graph, V is its set of vertices, and E is its set of edges:</em></p>
<ul>
    <li><em>Found a path from a vertex s to a vertex t</em></li>
    <li><em>Check that each edge traversed is in the set E</em></li>
    <li><em>If an edge not in E is found, return <b>FALSE</b></em></li>
    <li><em>For each vertex x in V</em></li>
    <li><em>If x is traversed more than once, return <b>FALSE</b></em></li>
    <li><em>If x is not traversed at all, return <b>FALSE</b></em></li>
    <li><em>return <b>True</b></em></li>
</ul>
The above algorithm takes polynomial time, and so Hamiltonian Path is in NP.
</p>

```python
import numpy as np
import random as rnd
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict
from numpy import random
```

```python
def HamiltonianPaths(G, n, visited, path, vertex, iterations):
    
    iterations[0]+=1
    if len(path) == n:
        return True
     
    for w in GetAdjacents(G,vertex):
        iterations[0]+=1
        if not visited[w]:
            
            visited[w] = True
            path.append(w)
 
            if HamiltonianPaths(G, n, visited, path, w, iterations):
                return True
 
            visited[w] = False
            path.pop()
            
    return False

def DeterministicHamiltonianPath(G):
    
    iterations = [0]
    n = len(G.nodes())
    for node in G.nodes:
        start_vertex = node
        visited = [False for i in range(n)]
        visited[start_vertex] = True 
        path = [start_vertex]
        if HamiltonianPaths(G, n, visited, path, start_vertex, iterations):
            return True, iterations[0]
 
    return False, iterations[0]

def NonDeterministicHamiltonianPath(G):
    
    path = list(G.nodes)
    random.shuffle(path)
        
    return(CheckHamiltonPath(path,G))

def CheckHamiltonPath(path, G):
    
    if len(path) < 1:
        return False
    
    for i in range (1, len(path)):
        if (path[i-1]) not in GetAdjacents(G,path[i]):
            return False
    
    return True

def GetAdjacents(G,vertex):
    adjacents = []
    for i in G.edges():
        if i[0] == vertex:
            adjacents.append(i[1])
        elif i[1] == vertex:
            adjacents.append(i[0])
    return adjacents
```

```python
def PrintGraph(G,be,bn,re=[],rn=[]):
    
    node_pos=nx.get_node_attributes(G,'pos')
    arc_weight=nx.get_edge_attributes(G,'weight')

    nx.draw_networkx_nodes(G, node_pos, nodelist=rn, cmap=plt.get_cmap('jet'),node_color = "#FF9A00", node_size = 500)
    nx.draw_networkx_nodes(G, node_pos, nodelist=bn, cmap=plt.get_cmap('jet'),node_color = "#cacaca", node_size = 500)
    nx.draw_networkx_labels(G, node_pos)
    nx.draw_networkx_edges(G, node_pos, edgelist=re, edge_color='#FF9A00', arrows=False)
    nx.draw_networkx_edges(G, node_pos, edgelist=be, edge_color='black', arrows=False)
    nx.draw_networkx_edge_labels(G, node_pos, edge_labels=arc_weight)
    plt.tight_layout()
    plt.axis('off')
    plt.show()
```

```python
G=nx.DiGraph()

G.add_node(0,pos=(0,0))
G.add_node(1,pos=(-1,-1))
G.add_node(2,pos=(0,-0.7))
G.add_node(3,pos=(1,-1))

list_arcs = [(0, 1),(0, 2),(0, 3),(1, 0),(1, 2),(1, 3),(2, 0),(2, 1),(2, 3),(3, 0),(3, 1),(3, 2)]
G.add_edges_from(list_arcs)

PrintGraph(G,G.edges(),G.nodes())
```

```python
print("Deterministc algorithm: " + str(DeterministicHamiltonianPath(G)[0]))
print("Nondeterministc algorithm: " + str(NonDeterministicHamiltonianPath(G)))
```

```python
x = [4,5,6,7,8]        # Different input sizes
samples = 100         # Quantity of samples to be averaged
hamiltoninan_it = [] # Quantity of iterations
```

```python
# Calculate the average iterations of all approaches
for i in range(len(x)):
    
    it1 = 0
    iterations = 0
    
    for k in range (0,samples):
        
        G1=nx.DiGraph()
        
        list_arcs = []
        for x1 in range(x[i]):
            create_edge = random.choice([True,False])
            if create_edge:
                for x2 in range(x[i]):
                    if x1 != x2:
                        list_arcs.append((x1,x2))
        
        G1.add_edges_from(list_arcs)
        
        solution, iterations = DeterministicHamiltonianPath(G1)
        
        it1 = it1 + iterations
    
    hamiltoninan_it.append(it1/samples)
```

```python
# Plots the complexity graph
plt.figure(figsize=(10, 8))
plt.plot(x,hamiltoninan_it, c="green", ls="-", lw = 1, marker="o", label="Hamiltonian Path")
plt.title("Hamiltonian Path Complexity",fontsize=14)
plt.xlabel("Input length",fontsize=14)
plt.ylabel("Number of iterations",fontsize=14)
plt.legend(fontsize=13)
plt.show()
```
