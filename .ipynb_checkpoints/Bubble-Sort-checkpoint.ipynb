```python
import random
import matplotlib.pyplot as plt
```

```python
# Returns the number of iterations of bubble sort (swaps and comparisons)
def bubbleSortIteration(arr):
    swaps = 0
    comparisons = 0
    for j in range(len(arr)-1,0,-1):
        for i in range(j):
            comparisons += 1 
            if arr[i]>arr[i+1]:
                swaps += 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
    return (comparisons + swaps)
```

```python
x = [10, 25, 50, 100, 200, 500] # Different input sizes
best_case = []                  # Quantity of iterations in the best case
worst_case = []                 # Quantity of iterations in the worst case
```

```python
# Calculate the iterations in best and worst case
for i in range(len(x)):
    best = []
    worst = []
    for j in range (x[i]):
        best.append(j)
        worst.append((j-x[i]+1)*-1)
    best_case.append(bubbleSortIteration(best))
    worst_case.append(bubbleSortIteration(worst))
```

```python
# Plots the complexity graph
plt.figure(figsize=(10, 8))
plt.plot(x,best_case, c="green", ls="-", lw = 1, marker="o", label="Best case")
plt.plot(x,worst_case, c="red", ls="-",  lw = 1, marker="o", label="Worst case")
plt.title("Bubble Sort iterations with different inputs",fontsize=14)
plt.xlabel("Input length",fontsize=14)
plt.ylabel("Number of iterations",fontsize=14)
plt.legend(fontsize=13)
plt.show()
```
