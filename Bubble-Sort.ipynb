```python
import matplotlib.pyplot as plt
```

```python
# Returns the number of iterations of bubble sort (swaps and comparisons)
def bubbleSortIteration(array):
    swaps = 0
    comparisons = 0
    arr = array.copy()
    for i in range(len(arr)):
        for j in range(len(arr) - 1):
            comparisons += 1
            if arr[j] > arr[j+1]:
                swaps += 1
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return (comparisons + swaps)

# Returns the number of iterations of improved bubble sort (swaps and comparisons)
def improvedBubbleSortIteration(array):
    swaps = 0
    comparisons = 0
    has_swapped = True
    arr = array.copy()
    while(has_swapped):
        has_swapped = False
        for i in range(len(arr) - 1):
            comparisons += 1
            if arr[i] > arr[i+1]:
                swaps += 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
                has_swapped = True
    return (comparisons + swaps)
```

```python
x = [10, 25, 50, 100, 200, 500] # Different input sizes
best_case_1 = []                 # Quantity of iterations - Bubble Sort best case
worst_case_1 = []                # Quantity of iterations - Bubble Sort worst case
best_case_2 = []                 # Quantity of iterations - Improved Bubble Sort best case
worst_case_2 = []                # Quantity of iterations - Improved Bubble Sort worst case
```

```python
# Calculate the iterations in best and worst case
for i in range(len(x)):
    best = []
    worst = []
    for j in range (x[i]):
        best.append(j)
        worst.append((j-x[i]+1)*-1)
    best_case_1.append(bubbleSortIteration(best))
    worst_case_1.append(bubbleSortIteration(worst))
    best_case_2.append(improvedBubbleSortIteration(best))
    worst_case_2.append(improvedBubbleSortIteration(worst))
    
```

```python
# Plots the complexity graph
plt.figure(figsize=(10, 8))
plt.plot(x,best_case_1, c="red", ls="-.", lw = 1, marker="^", ms=10, label="Buble Sort best case")
plt.plot(x,worst_case_1, c="red", ls="-.",  lw = 1, marker="^", ms=10, label="Buble Sort worst case")
plt.plot(x,best_case_2, c="blue", ls="-.", lw = 1, marker="v", ms=10, label="Improved Buble Sort best case")
plt.plot(x,worst_case_2, c="blue", ls="-.",  lw = 1, marker="v", ms=10, label="Improved Buble Sort worst case")
plt.title("Bubble Sort vs Improved Buble Sort iterations with different inputs",fontsize=14)
plt.xlabel("Input length",fontsize=14)
plt.ylabel("Number of iterations",fontsize=14)
plt.legend(fontsize=13)
plt.show()
```
