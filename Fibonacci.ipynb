```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
# Dynamic Programming approach of Fibonacci
def DynamicFibonacci(n, it):
    if (n <= 0):
        return 0
  
    fibo = np.zeros(n+1, dtype=int) #fibo[0] will be 0
    fibo[1] = 1
  
    # Calculate fibonnaci
    for i in range(2,n+1):
        it[0]+=1
        fibo[i] = fibo[i-1] + fibo[i-2]
         
    return fibo[n]

# Brute Force approach of Fibonacci
def BruteFibonacci(n, it):
    if (n <= 0):
        return 0
    if (n == 1):
        return 1
    
    it[0]+=1
    return (BruteFibonacci(n-2,it) + BruteFibonacci(n-1,it))
```

```python
x = [10, 12, 14, 16, 18, 20] # Different inputs
bruteFibo = []               # Quantity of iterations - Brute Force approach
dynamicFibo = []             # Quantity of iterations - Dynamic Programming approach
```

```python
# Calculate the iterations of all approaches
for i in range(len(x)):
    iterations1, iterations2 = [0],[0]
    
    BruteFibonacci(x[i], iterations1)
    DynamicFibonacci(x[i], iterations2)
        
    bruteFibo.append(iterations1[0])
    dynamicFibo.append(iterations2[0])
```

```python
# Plots the complexity graph
plt.figure(figsize=(10, 8))
plt.plot(x,bruteFibo, c="red", ls="-", lw = 1, marker="o", label="Brute Force approach")
plt.plot(x,dynamicFibo, c="green", ls="-",  lw = 1, marker="o", label="Dynamic Programming approach")
plt.title("Brute Force Fibonacci vs Dynamic Programming Fibonacci",fontsize=14)
plt.xlabel("Input",fontsize=14)
plt.ylabel("Number of iterations",fontsize=14)
plt.legend(fontsize=13)
plt.show()
```
