```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
# Function to do improved quick sort  
def ImprovedQuicksort(A, l, r, iterations):
    
    iterations[0] += 1
    if l < r:
        iterations[0] += 1
        if (r-l + 1 < 16):
            InsertionSort(A, l, r+1, iterations);
        else:
            s = HoarePartition(A, l, r, iterations)
            ImprovedQuicksort(A, l, s-1, iterations)
            ImprovedQuicksort(A, s+1, r, iterations)

# Function to do quick sort    
def Quicksort(A, l, r, iterations):
    
    iterations[0] += 1
    
    if l < r:
        s = HoarePartition(A, l, r, iterations)
        Quicksort(A, l, s-1, iterations)
        Quicksort(A, s+1, r, iterations)

# Function used in Quick Sorte to make partitions
def HoarePartition(A, l, r, iterations):
    
    p = A[l]
    i = l
    j = r + 1
    
    while True:
        while i < r:
            i += 1
            iterations[0] += 1
            if(A[i] >= p):
                break
        while j > l:
            j -= 1
            iterations[0] += 1
            if(A[j] <= p):
                break
                
        A[i], A[j] = A[j], A[i]
        
        iterations[0] += 1
        if(i >= j):
            break
    
    A[i], A[j] = A[j], A[i]
    A[l], A[j] = A[j], A[l]

    return j

# Function to do insertion sort
def InsertionSort(A, l, r, iterations):
  
    for i in range(l, r):
        
        v = A[i]
        j = i-1
        iterations[0] += 1
        
        while j >=l and A[j] > v:
            
            iterations[0] += 1
            A[j+1] = A[j]
            j -= 1
            
        A[j+1] = v
```

```python
x = [10, 25, 50, 100, 200, 500] # Different input sizes
samples = 100                   # Quantity of samples to be averaged
quick_it = []                   # Quantity of iterations - Quick Sort
improved_it = []                # Quantity of iterations - Improved Quick Sort
```

```python
# Calculate the average iterations of all approaches
for i in range(len(x)):
    
    it1, it2 = 0,0
    iterations1, iterations2 = [0],[0]
    
    for k in range (0,samples):
        
        array = random.randint(100, size=(x[i]))
        array2 = array.copy()
        
        Quicksort(array, 0, len(array) - 1, iterations1)
        ImprovedQuicksort(array2, 0, len(array2) - 1, iterations2)
        
        it1 = it1 + iterations1[0]
        it2 = it2 + iterations2[0]
        
    quick_it.append(it1/samples)
    improved_it.append(it2/samples)
```

```python
# Plots the complexity graph
plt.figure(figsize=(10, 8))
plt.plot(x,quick_it, c="green", ls="-", lw = 1, marker="o", label="Quick Sort")
plt.plot(x,improved_it, c="orange", ls="-",  lw = 1, marker="o", label="Improved Quick Sort")
plt.title("Complexity analysis of Quick Sort and Improved Quick Sort",fontsize=14)
plt.xlabel("Input length",fontsize=14)
plt.ylabel("Number of iterations",fontsize=14)
plt.legend(fontsize=13)
plt.show()
```
